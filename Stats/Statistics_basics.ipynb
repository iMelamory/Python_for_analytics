```python
#Выбросы для пульса
q1 = data_filtered['pulse'].quantile(0.25)
q3 = data_filtered['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = data_filtered[data_filtered['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
new = pd.concat([data_filtered, remove_outliers]).drop_duplicates(keep=False)
new.iloc[:, [1,2,3]].dropna()
```

```python
#Выбросы для rectal_temp
q1 = data_filtered['rectal_temperature'].quantile(0.25)
q3 = data_filtered['rectal_temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = data_filtered[data_filtered['rectal_temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature')
new = pd.concat([data_filtered, remove_outliers]).drop_duplicates(keep=False)
new.iloc[:, [2]].dropna()
```

```python
#Выбросы для respiratory_rate
q1 = data_filtered['respiratory_rate'].quantile(0.25)
q3 = data_filtered['respiratory_rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = data_filtered[data_filtered['respiratory_rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory_rate')
new_1 = pd.concat([data_filtered, remove_outliers]).drop_duplicates(keep=False)
new_1.iloc[:, [4]].dropna()
```

```python
df = pd.read_csv('C:\\Users\\Irina\\Documents\\Материалы PYDA\\Stats\\практика_и_дз\\horse_data.csv',
                   names = ['surgery', 'age', 'Hospital_Number', 'rectal_temperature', 'pulse', 'respiratory_rate', 'temperature_of_extremities', 'peripheral pulse', 'mucous membranes', 'capillary refill time', 'pain', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination', 'abdomen', 'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome', 'other', 'other1', 'other2', 'other3', 'other4'], na_values=['?']).iloc[:, [0,1,3,4,5,6,10,22]]

df
```

```python
df['rectal_temperature'].fillna(df['rectal_temperature'].median(), inplace=True)
df['pain'].fillna(3, inplace=True)
df['temperature_of_extremities'].fillna(df['temperature_of_extremities'].median(), inplace=True)
df['respiratory_rate'].fillna(df[(df.surgery == 1) & (df.age == 1)].respiratory_rate.median(), inplace=True)
df['respiratory_rate'].fillna(df[(df.surgery == 1) & (df.age == 9)].respiratory_rate.median(), inplace=True)
df['respiratory_rate'].fillna(df[(df.surgery == 2) & (df.age == 9)].respiratory_rate.median(), inplace=True)
df['respiratory_rate'].fillna(df[(df.surgery == 2) & (df.age == 1)].respiratory_rate.median(), inplace=True)
```
